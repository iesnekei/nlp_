```python
import pandas as pd
```

```python
data = open('corpus.txt').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
trainDF.label.nunique()
```

```python
from sklearn import model_selection, preprocessing, linear_model

train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
import numpy as np
```

```python
def testing_mode(model_, vectorizer_type, analyzer_, token_pattern_, n_features_):

    count_vect = vectorizer_type(analyzer=analyzer_, token_pattern=token_pattern_, n_features=n_features_)
    count_vect.fit(trainDF['text'])

    xtrain_count =  count_vect.transform(train_x)
    xvalid_count =  count_vect.transform(valid_x)

    classifier = model_
    classifier.fit(xtrain_count, train_y)

    predictions = classifier.predict(xvalid_count)
    predictions = np.where(predictions > 0.5,1,0)
    print(predictions)
    results['model'].append(type(model_).__name__)
    results['vectorization_method'].append(vectorizer_type)
    results['parameters'].append({'analyzer': analyzer_,
                                  'token_pattern': token_pattern_})
    results['n_features'].append(n_features_)
    results['score'].append(accuracy_score(valid_y, predictions))


```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D, Conv1D, GRU, LSTM, Dropout
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization
```

```python
train_data = tf.data.Dataset.from_tensor_slices((train_x, train_y))
valid_data = tf.data.Dataset.from_tensor_slices((valid_x, valid_y))
```

```python
train_data = train_data.batch(16)
valid_data = valid_data.batch(16)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_data = train_data.cache().prefetch(buffer_size=AUTOTUNE)
valid_data = valid_data.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
def custom_standardization(input_data):
    
    return input_data

vocab_size = 10000
seq_len = 100

vectorize_layer = TextVectorization(
    standardize=custom_standardization,
    max_tokens=vocab_size,
    output_mode='int',
    output_sequence_length=seq_len)

# Make a text-only dataset (no labels) and call adapt to build the vocabulary.
text_data = train_data.map(lambda x, y: x)
vectorize_layer.adapt(text_data)
```

```python
embedding_dim=200

model = Sequential([
    vectorize_layer,
    Embedding(vocab_size, embedding_dim),
    Conv1D(200, (3)),
    Conv1D(200, (2)),
    GRU(300),
    #GlobalAveragePooling1D(),
    Dense(100, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(train_data, validation_data=valid_data, epochs=15)
```
